```python
data = pd.read_csv('data/Pokemon.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
# plotting line graphs
plt.figure(figsize=(18,10))
data.Speed.plot(kind = 'line', color = 'g', linewidth =1, label = 'Speed', alpha=0.5, grid = True, linestyle = ':')
data.Defense.plot(color='r', label = 'Defense', linewidth = 1, alpha = 0.5, grid = True, linestyle = '-')
plt.legend (loc = 'upper right')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Speed of Pokemon vs Defense')
plt.show()
```

# Scatter Plot

```python
data.plot(kind = 'scatter', x = 'Attack', y = 'Defense', color = 'r', alpha = 0.5)

plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack Defense scatter plot')
```

```python
# Histogram

data.Speed.plot(kind = 'hist', bins =50, figsize = (12,12))
plt.show()
```
